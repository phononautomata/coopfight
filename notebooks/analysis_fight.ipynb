# COOPFIGHT: FIGHTING EVENTS ANALYSIS

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def csf_tullock(resources_focal, resources_enemy, technology):
    x = np.power(resources_focal, technology)
    y = np.power(resources_enemy, technology)
    return x / (x + y)
```

```python
resources_ratio = np.linspace(0.0, 10.0, 1000)
resources_ref = 1.0

list_technology = [0.0, 0.01, 0.1, 0.25, 0.5, 0.75, 0.9, 1.0, 1.5, 5.0, 10.0]
```

```python
fig = plt.figure(figsize=(18, 9))

for technology in list_technology:
    plt.plot(resources_ratio, csf_tullock(resources_ref, resources_ratio, technology))

plt.show()
```
